# Drop rows with NaN values

```python
nba = pd.read_csv('datasets/nba.csv')
nba.head()
```

```python
nba.dropna().head()
```

```python
nba.dropna(how='all').head()
```

```python
nba.dropna(subset=['Salary']).head()
```

```python
nba.dropna(how='all', inplace=True)
nba.dropna(axis='columns').head()
```
